```python
import numpy as np
import matplotlib.pyplot as plt

N = int(1e6)

mu = 10.2 # sample mean; input positive real number
sigma = 6.7 # sample std

m = 2.4 # slope
b = 104.1 # Intercept
```

```python
x = np.random.normal(mu, sigma, N)

y = m*x + b + sigma * np.random.standard_normal(N) # regression 
```

```python
plt.plot(x,y,'k.')
```

```python
x.mean()
```

```python
x.std
```

```python
x.var()
```

```python
y.mean()
```

```python
y.std()
```

```python
# residuals = distance from point to regression line. Can be in both x and y direction 
(y - y.mean()).mean() # mean of residuals 
```

```python
((y-y.mean())/y.std()).std() 
```

```python
x_hist, bin_edges = np.histogram(y, bins = 20)
bin_center = (bin_edges[0:-1] + bin_edges[1:]) / 2 # [0:-1] = go to end and scooch back one; [1:] = go to end 

plt.bar(bin_center, x_hist, width=0.9*(bin_center[1]-bin_center[0]))
```

```python
# test correlation 

np.corrcoef(x,y) # gives whole correlation matrix
```
