```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0.0,2*np.pi,1000)
```

```python
sint = np.sin(t)
```

```python
cost = np.cos(t)
```

```python
t.size # says how many individual elements in the array 
t.shape # says what the shape of t is
```

```python
plt.figure(figsize=(12,6))
plt.plot(t,cost,label='cos (t)')
plt.plot(t,sint,label='sin (t)')
plt.xlabel('Time [hr]')
plt.ylabel('Wave Height [m]')
plt.title('Tidal Wave Heights at Fraggle Rock Cove')
plt.grid(axis='y')
plt.legend(loc='lower right')
plt.show()
```
